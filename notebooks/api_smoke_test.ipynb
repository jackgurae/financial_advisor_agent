# API Smoke Test

This notebook smoke-tests the live integrations used by the app:

- Financial Modeling Prep via `app.tools.get_symbol_data`
- Google News via `app.tools.get_news`

Run the cells from top to bottom.


```python
import json
import os
import sys
from pathlib import Path

from dotenv import load_dotenv

project_root = (
    Path.cwd().resolve().parent
    if Path.cwd().name == "notebooks"
    else Path.cwd().resolve()
)
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

load_dotenv(project_root / ".env")

print("Project root:", project_root)
print("OPENAI_API_KEY set:", bool(os.getenv("OPENAI_API_KEY")))
print("FMP_API_KEY set:", bool(os.getenv("FMP_API_KEY")))
```

```python
from app.tools import get_symbol_data

symbol = "AAPL"
raw_symbol_data = get_symbol_data(symbol)
symbol_data = json.loads(raw_symbol_data)

print("Top-level keys:", list(symbol_data.keys()))
print("Profile company name:", symbol_data.get("profile", {}).get("companyName"))
print("Quote price:", symbol_data.get("quote", {}).get("price"))
print("Valuation keys:", list(symbol_data.get("valuation", {}).keys())[:10])
```

```python
from app.tools import get_news

news = json.loads(get_news("AAPL"))

print("News item count:", len(news))
if news:
    first_item = news[0]
    print("First title:", first_item.get("title"))
    print("First publisher:", first_item.get("publisher", {}).get("title"))
else:
    print("No news items returned.")
```
